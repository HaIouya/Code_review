```python
from collections import defaultdict
import pandas as pd
# 创建一个空的字典来存储每个职位的技术栈
job_skills_dict = defaultdict(list)
csv_path = 'Boss直聘_skills.csv'  # 假定文件已上传到这个路径
df = pd.read_csv(csv_path)

# 提取技术栈列中的每个技术，并转换为 JSON 格式
for index, row in df.iterrows():
    skills_str = row['技术栈']
    skills_list = [skill.split('. ', 1)[1].strip() if '. ' in skill else skill.strip() for skill in skills_str.split('\n') if skill.strip()]  # 清除空白字符并分割技术
    job_skills_dict[row['职位名称']].append(skills_list)

# 将技术列表转换为 JSON 格式，并统计每个技术的出现频率
job_skills_json = {}
for job_title, skills_list in job_skills_dict.items():
    skill_counts = defaultdict(int)
    for skills in skills_list:
        for skill in skills:
            skill_counts[skill] += 1
    job_skills_json[job_title] = dict(skill_counts)

# 预览处理后的数据
job_skills_json
```

```python
from collections import defaultdict
import pandas as pd
# 创建一个空的字典来存储每个职位的技术栈
job_skills_dict = defaultdict(list)
csv_path = 'Boss直聘_skills.csv'  # 假定文件已上传到这个路径
df = pd.read_csv(csv_path)

# 提取技术栈列中的每个技术，并转换为 JSON 格式
for index, row in df.iterrows():
    skills_str = row['工作内容']
    skills_list = [skill.split('. ', 1)[1].strip() if '. ' in skill else skill.strip() for skill in skills_str.split('\n') if skill.strip()]  # 清除空白字符并分割技术
    job_skills_dict[row['职位名称']].append(skills_list)

# 将技术列表转换为 JSON 格式，并统计每个技术的出现频率
job_skills_json = {}
for job_title, skills_list in job_skills_dict.items():
    skill_counts = defaultdict(int)
    for skills in skills_list:
        for skill in skills:
            skill_counts[skill] += 1
    job_skills_json[job_title] = dict(skill_counts)

# 预览处理后的数据
job_skills_json
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
from matplotlib.font_manager import FontProperties
from collections import Counter

# 用户输入或选择的职业
selected职业 = '数仓工程师'

# 提取指定职业的技能列表
skills = job_skills_json.get(selected职业, {})

# 对技能列表按计数进行排序，并选择计数最高的六个技能
top_skills = [skill for skill, count in Counter(skills).most_common(6)]

# 如果技能不足六个，只显示现有的技能
num_skills = len(top_skills)
labels = top_skills
sizes = [skills[skill] for skill in top_skills]
colors = plt.cm.viridis.colors[:num_skills]


chinese_font = FontProperties(fname='/usr/share/fonts/MyFonts/simhei.ttf')


plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title(f"Top Skills for '{selected职业}'", fontsize=16)
plt.legend(title='Skills', bbox_to_anchor=(1, 0, 0.5, 1), loc="center right", fontsize=12)
plt.show()
```
