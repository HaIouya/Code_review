## 技术栈

### json封装

```python
from collections import defaultdict
import pandas as pd
# 创建一个空的字典来存储每个职位的技术栈
job_skills_dict = defaultdict(list)
csv_path = 'Boss直聘_skills.csv'  # 假定文件已上传到这个路径
df = pd.read_csv(csv_path)

# 提取技术栈列中的每个技术，并转换为 JSON 格式
for index, row in df.iterrows():
    skills_str = row['技术栈']
    skills_list = [skill.split('. ', 1)[1].strip() if '. ' in skill else skill.strip() for skill in skills_str.split('\n') if skill.strip()]  # 清除空白字符并分割技术
    job_skills_dict[row['职位名称']].append(skills_list)

# 将技术列表转换为 JSON 格式，并统计每个技术的出现频率
job_skills_json = {}
for job_title, skills_list in job_skills_dict.items():
    skill_counts = defaultdict(int)
    for skills in skills_list:
        for skill in skills:
            skill_counts[skill] += 1
    job_skills_json[job_title] = dict(skill_counts)

# 预览处理后的数据
job_skills_json
```

### 相似合并

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# 初始化Sentence-BERT模型
model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")

# 假设这是您要处理的数据
# 示例数据结构：{'job_name': {'skill': num, 'skill2': num}, 'job_name2': {'skill': num, 'skill2': num}}
data = job_skills_json

# 遍历每个职业
for job_name, skills in data.items():
    # 获取每个技能的名称
    skill_names = list(skills.keys())
    # 获取每个技能的Sentence-BERT嵌入
    skill_embeddings = model.encode(skill_names)

    # 计算相似度矩阵
    similarities = model.similarity(skill_embeddings, skill_embeddings)

    # 找出相似度超过0.85的技能并进行合并
    threshold = 0.85
    merged_skills = set()

    for i, skill1 in enumerate(skill_names):
        for j, skill2 in enumerate(skill_names):
            if j > i:  # 只比较skill1后面的技能
                score = similarities[i][j]
                if score > threshold:
                    # 如果相似度超过阈值，合并数据
                    if skill1 not in merged_skills:
                        data[job_name][skill1] += data[job_name][skill2]
                        merged_skills.add(skill2)

    # 移除已合并的技能
    for skill in merged_skills:
        del data[job_name][skill]

# 输出更新后的数据
data
```

### 星图绘制

```python
from ipywidgets import Dropdown, Button, VBox, Output, interact
import matplotlib
import matplotlib.pyplot as plt

# 修改plot_radar_chart_corrected函数，根据职业的技能数量动态调整N值
def plot_radar_chart_dynamic(selected_job_data):
    # 创建一个图形和一组子图
    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

    # 计算每个职业的轴位置
    angles = np.linspace(0, 2 * np.pi, len(selected_job_data), endpoint=False).tolist()

    # 绘制每个职业的星图
    for i, (job, skills) in enumerate(selected_job_data.items()):
        # 仅考虑该职业的所有技能
        top_skills = dict(skills)
        values = list(top_skills.values())
        values += values[:1]
        angles_job = angles[:len(top_skills)] + angles[:1]
        ax.fill(angles_job, values, color=np.random.rand(3).tolist(), alpha=0.25)
        ax.plot(angles_job, values, color=np.random.rand(3).tolist(), linewidth=2, linestyle='solid', label=job)

    # 设置图表的标签
    ax.set_yticklabels([])
    ax.set_xticks(angles)
    ax.set_xticklabels([f"{key}" for key in top_skills.keys()])

    # 添加图例
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

    # 显示图表
    plt.show()

# 创建一个交互式下拉菜单来选择职业
def plot_radar_for_selected_job(normalized_data, N=5):
    # 创建一个输出对象来显示图表
    out = Output()

    # 创建一个下拉菜单
    job_dropdown = Dropdown(options=list(normalized_data.keys()))

    # 创建一个按钮来触发绘图
    plot_button = Button(description="绘制星图")

    # 定义按钮的点击事件
    def on_button_clicked(b):
        with out:
            out.clear_output()
            # 绘制所选职业的星图
            selected_job = job_dropdown.value
            if selected_job:
                plot_radar_chart_corrected({selected_job: normalized_data[selected_job]}, N=N)

    # 将点击事件绑定到按钮
    plot_button.on_click(on_button_clicked)

    # 显示菜单和按钮
    display(VBox([job_dropdown, plot_button, out]))

plot_radar_for_selected_job_dynamic = interact(plot_radar_chart_dynamic, selected_job_data=normalized_data)
plot_radar_for_selected_job_dynamic

```
