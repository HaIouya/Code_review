```python

# 初始化一个字典用于存储 {市: 省} 的对应关系
city_to_province = {}

# 遍历数据结构
for province_id, province_data in data.items():
    
    if isinstance(province_data, dict)  and province_id > 0:
        for city_id, city_name in province_data.items():
            
            if isinstance(city_name, str):
                
                province_name = data[0][province_id]
                city_to_province[city_name] = province_name

# 将转换后的数据保存为 JSON 文件
with open('city_to_province.json', 'w', encoding='utf-8') as file:
    json.dump(city_to_province, file, ensure_ascii=False, indent=2)

print("转换完成，结果已保存到 city_to_province.json 文件中。")
```
